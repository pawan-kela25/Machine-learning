```python
from sklearn.datasets import load_iris
data = load_iris()
```

```python
import pandas as pd
df = pd.DataFrame(data = data.data,columns=data.feature_names)
df['target'] = data.target

X = df.drop('target',axis=1)
y = df['target']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.svm import SVC
model = SVC(C=10,kernel='poly',gamma='auto')
model.fit(X_train,y_train)
model.predict(X_test)
model.score(X_test,y_test)
```

```python
help(model)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.svm import SVC
cross_val_score(SVC(C=1,kernel='linear',gamma='auto'),data.data,data.target)
```

```python
from sklearn import svm
from sklearn.model_selection import GridSearchCV
clf = GridSearchCV(svm.SVC(gamma='auto'),{
    'C':[1,10,20],
    'kernel':['linear','rbf']
},cv=5,return_train_score=False)
clf.fit(X_train,y_train)
clf.cv_results_
```

```python
df2 = pd.DataFrame(clf.cv_results_)
df2.drop(['mean_fit_time','std_fit_time','mean_score_time','std_score_time','split0_test_score','split1_test_score','split2_test_score','split3_test_score','split4_test_score'],axis=1,inplace=True)
```
