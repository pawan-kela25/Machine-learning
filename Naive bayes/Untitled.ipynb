```python
import pandas as pd
df = pd.read_csv('titanic.csv')
df.head()
```

```python
df.drop(['PassengerId','Name','Ticket','Cabin','Parch','SibSp','Embarked'],axis='columns',inplace=True)
```

```python
df.head()
```

```python
df2 = pd.get_dummies(df['Sex'])
df2 = df2.replace({False:0,True:1})
df2.head()
```

```python
df = pd.concat([df,df2],axis='columns')
df
```

```python
df = df.drop('Sex',axis='columns')
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df["Age"].mean(),inplace=True)
```

```python
df.drop("female",inplace=True,axis='columns')
```

```python
df.head()
```

```python
inputs = df.drop('Survived',axis='columns')
target = df['Survived']
inputs
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(inputs,target,test_size=0.2)

from sklearn.naive_bayes import GaussianNB
model = GaussianNB()

model.fit(X_train,y_train)
model.predict(X_test)
model.score(X_test,y_test)
```
