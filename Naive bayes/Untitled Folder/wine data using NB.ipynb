```python
from sklearn.datasets import load_wine
data = load_wine()
```

```python
dir(data)
```

```python
import pandas as pd
df = pd.DataFrame(data = data.data,columns=data.feature_names)

df['target'] = data.target
df.shape
```

```python
df.head()
```

```python
inputs = df.drop('target',axis=1)
target = df['target']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs, target, test_size=0.33, random_state=42)
```

```python
from sklearn.naive_bayes import MultinomialNB
model1 = MultinomialNB()
model1.fit(X_train,y_train)
model1.score(X_test,y_test)
```

```python
df['alcohol'].unique()
```

```python
from sklearn.naive_bayes import GaussianNB
model2 = GaussianNB()

model2.fit(X_train,y_train)
model2.score(X_test,y_test)

from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,model2.predict(X_test))
```

```python
help(confusion_matrix)
```

```python
from sklearn.pipeline import Pipeline
clf = Pipeline([('model1',MultinomialNB()),('model2',GaussianNB())])
clf.fit(X_train,y_train)
clf.score(X_test,y_test)
```
