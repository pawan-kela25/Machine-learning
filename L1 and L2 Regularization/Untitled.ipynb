```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
df = pd.read_csv('Melbourne_housing_FULL.csv')
df.head()
```

```python
df.drop(['Date','YearBuilt','Lattitude','Longtitude','Address','Postcode'],inplace=True,axis=1)
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.dropna(subset = ['CouncilArea','Regionname','Propertycount'],axis=0,inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
cols_to_fill = ['Bathroom','Landsize','Price']
df['Car'].fillna(0,inplace=True)
df['Bathroom'].fillna(0,inplace=True)
df['Landsize'].fillna(0,inplace=True)
df['Price'].fillna(df['Price'].mean(),inplace=True)
df['Bedroom2'].fillna(0,inplace=True)
df['BuildingArea'].fillna(0,inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df.drop(['Suburb','Type',"Method",'SellerG','CouncilArea','Regionname'],axis=1,inplace=True)
```

```python
X = df.drop('Price',axis=1)
y = df['Price']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)

print(f'train model score : {model.score(X_train,y_train)}')
print(f'test model score : {model.score(X_test,y_test)}')
```
