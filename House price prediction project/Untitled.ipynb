```python
import pandas as pd
df = pd.read_csv('bengaluru_house_prices.csv')
df.head()
```

```python
df2 = df.drop(['area_type','availability','society'],axis=1)
df2 = df2.drop('balcony',axis=1)
df2.head()
```

```python
df2.shape
```

```python
df2.isnull().sum()
```

```python
df3 = df2.dropna()
df3.head()
```

```python
df3.head()
```

```python
df3['bhk'] = df3['size'].apply(lambda x:int(x.split(' ')[0]))
df3.head()
```

```python
df4 = df3.drop('size',axis=1)
df4.head()
```

```python
df4['total_sqft'].unique()
```

```python
def convert_range_to_float(x):
    token = x.split("-")
    if(len(token)==2):
        return (float(token[0])+float(token[1]))/2
    try:
        return float(x)
    except:
        return None
    
    
df4['total_sqft'] = df4['total_sqft'].apply(convert_range_to_float)
```

```python
def is_float(x):
    try:
        return float(x)
    except:
        return x

    
df4['total_sqft']  = df4['total_sqft'].apply(is_float)
df4['total_sqft'].unique()
```
